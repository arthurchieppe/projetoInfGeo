```python
pd.set_option('display.max_rows', 50)
pd.set_option('display.max_columns', None)
```

```python
ms = pd.read_csv("MS.csv")
```

```python
ms
```

```python
# Drop columns that rows contain the string "Total"
ms = ms.loc[:, ~(ms == 'Total').any()]
```

```python
# Replace "-" with 0
ms = ms.replace('-', 0)

```

```python
ms.shape
```

```python
df = ms
```

```python
df = df.iloc[:,2:-2]
df = df.replace('-', 0)
```

```python
df = df.apply(lambda x: pd.to_numeric(x))
```

```python
# temp = df.mean(axis=1)
df.info(verbose=True)
```

```python
# If row value is less than mean, set to 0, else set to 1
df = df.apply(lambda x: np.where(x < x.mean(), 0, 1))
```

```python
df
```

```python
# Drop all rows
temp = df.drop(df.index[:,])
```

```python
# Remove accents from columns and output to Keiya
temp.columns = temp.columns.str.normalize('NFKD').str.encode('ascii', errors='ignore').str.decode('utf-8')
temp
temp.to_csv("colunas.csv", index=False)
```
